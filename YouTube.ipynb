# 1. Find out the top 5 categories with maximum number of videos uploaded.

```python
a = pd.value_counts(yt['category'])    #using count fun 
b = a.head(5)                          #as we need top 5
print('Top 5 Categories with maximun no of videos uploaded are,\n',b)
```

# 2. Find out the top 10 rated videos.

```python
a = yt.groupby('raters')['rating'].mean()        #averaging the rating by no of persons giving rating and their rate
b = a.sort_values(ascending = False)              
b.head(10)
```

```python
aa = yt['id']                                         #for concatinating by video id
df = pd.concat([aa,a], axis = 1, sort = False)        #here we r concating
b = df.sort_values(['rating'],ascending = False)    
print('Top 10 rated videos are as, \n')
print(b.head(10))
```

# 3. Find out the most viewed videos.

```python
mx = yt['views'].max()             #using max function
a = yt['views'].idxmax()

video_id = list()
for i in yt['id']:
    video_id.append(i)

print("Most viewed video is of id {} and it have {} views.".format(video_id[a],mx))
```
